```python
import re

# Function to remove digits and GB
def extract_name(text):
    if pd.isna(text):
        return None
    # Remove numbers + GB + symbols
    text = re.sub(r'[\d/]+Gb', '', text)   # remove "256GB" etc
    text = re.sub(r'[^A-Za-z\s\-]', '', text) # remove symbols
    return text.strip()

names_only = clean.apply(extract_name)
print(names_only)
```

| Task                  | Pandas Method / Regex              |
| --------------------- | ---------------------------------- |
| Remove numbers        | `str.replace(r'\d+', '')`          |
| Remove GB             | `str.replace(r'\d+GB', '')`        |
| Remove symbols        | `str.replace(r'[^A-Za-z\s]', '')`  |
| Extract first letters | `str.extract(r'^([A-Za-z\s\-]+)')` |
| Apply custom function | `Series.apply(func)`               |


Create a clean table with title "Brand", "Model", "Storage"

```python
colors = ["Black", "White", "Blue", "Red", "Green", "Silver", "Gold"]
color_pattern = "|".join(colors)
```

```python
def extract_info(title):
    if pd.isna(title):
        return pd.Series([None, None, None])  # brand, model, storage
    
    # Extract brand (first word)
    brand = re.match(r'^[A-Za-z]+', title)
    brand = brand.group(0) if brand else None
    
    # Extract storage (number followed by GB)
    storage_matches = re.findall(r'\d+Gb', title)
    storage = storage_matches[-1] if storage_matches else None  # take last if multiple
    
    # Extract model (everything between brand and storage)
    if brand and storage:
        # Remove brand and storage from title
        model = title
        model = re.sub(r'^' + re.escape(brand) + r'\s*', '', model)  # remove brand
        model = re.sub(re.escape(storage), '', model)  # remove storage
        model = re.sub(r'\d+Gb/?', '', model)
        model = re.sub(color_pattern, '', model, flags=re.IGNORECASE)
        model = model.strip(" -")  # clean leftover dash or spaces
        model = model.replace('()', '')
        model = model.strip()
    else:
        model = None
    
    return pd.Series([brand, model, storage])
```

```python
df = clean.apply(extract_info)
df.columns = ["Brand", "Model", "Storage"]
print(df)
```
