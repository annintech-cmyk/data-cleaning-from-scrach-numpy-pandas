📌 Syntax:
add_subplot(rows, columns, position)

(4, 1, 1)

👉 Break it down:

1st position

```python
fig = plt.figure()
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)
plt.show()
```

🧠 Final takeaway
figure() → creates canvas
add_subplot(r,c,i) → divides into grid
ax → individual plot
show() → displays everything

👉 So total:
5 plots × 12 data points each = 60 data points visualized
[Plot 1 → first 12 months]
[Plot 2 → next 12 months]
[Plot 3 → next 12 months]
[Plot 4 → next 12 months]
[Plot 5 → next 12 months]

-Break large data into smaller chunks
-Compare trends over time
-Avoid clutter in one big graph

```python
fig = plt.figure(figsize=(12,12))

for i in range(5):
    ax = fig.add_subplot(5,1, i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['UNRATE'])

plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month

fig = plt.figure(figsize=(6,3))

plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['UNRATE'], c='red')
plt.plot(unrate[12:24]['MONTH'],  unrate[12:24]['UNRATE'], c='green')
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
colors = ['red','blue','green','orange','black']

for i in range(5):
    start_index = i * 12
    end_index = (i + 1) * 12  

    subset = unrate.iloc[start_index:end_index]

    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i])

plt.show()

```

```python
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['UNRATE'], c='red', label='1948')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['UNRATE'], c='blue', label='1949')

plt.legend(loc='upper left')
```
