Write a function to calculate total alcohol consumption of each county in each year

```python
world_alcohol_4 = world_alcohol.copy()
```

```python
def _row_county_year(country, year):
    is_country_year = (world_alcohol_4[:,2] == country) & (world_alcohol_4[:,0] == year)
    rows_with_country = world_alcohol_4[is_country_year]
    return rows_with_country

def calculate_alcohol_consumption(country, year):
    country_year_data = _row_county_year(country=country, year=year)
    country_alcohol = country_year_data[:,4]

    empty_strings = country_alcohol ==''
    country_alcohol[empty_strings] = "0"

    country_alcohol = country_alcohol.astype(float)

    total_country_drinking = country_alcohol.sum()

    return total_country_drinking

    
   
```

```python
print(calculate_alcohol_consumption("Canada","1984"))
```
