# Find number of people survived in each class

```python
classes = [1, 2, 3]
def total_survival_data():
    survived_num_group1 = 0
    survived_num_group2 = 0
    survived_num_group3 = 0
    for i, row in titianic_data.iterrows():
        
        if row["pclass"] == classes[0] and row["survived"] == 1:
            survived_num_group1 += 1
            
        elif row["pclass"] == classes[1] and row["survived"] == 1:
            survived_num_group2 += 1
            
        elif row["pclass"] == classes[2] and row["survived"] == 1:
            survived_num_group3 += 1

    print(f'Class 1: {survived_num_group1}')
    print(f'Class 2: {survived_num_group2}')
    print(f'Class 3: {survived_num_group3}')
```

```python
total_survival_data()
```

 This is a Pandas function used to summarize data (like Excel pivot tables)
 index="pclass" - Groups data by passenger class (pclass) -  Example values: 1, 2, 3
 values="fare" - The column you want to analyze - Here: ticket fare
 aggfunc=numpy.mean - The operation to apply on grouped data - Here: average (mean)

 1. Basic Syntax
 df.pivot_table(index, values, aggfunc)

2. Important Parameters

```python
import numpy
class_fares = titianic_data.pivot_table(index="pclass", values="fare", aggfunc=numpy.mean)
print(class_fares)
```

3. Multiple grouping

```python
class_fares_with_sex_pclass = titianic_data.pivot_table(index=["pclass", "sex"], values="fare", aggfunc="mean")
print(class_fares_with_sex_pclass)
```

4. Multiple values

```python
class_fares_and_age_with_pclass  = titianic_data.pivot_table(index="pclass", values=["fare", "age"], aggfunc="mean")
print(class_fares_and_age_with_pclass)
```

🔹 5. Columns parameter (like Excel pivot columns)

```python
spec = titianic_data.pivot_table(index="pclass", columns="sex", values="fare", aggfunc="mean")
print(spec)
```

6. Fill missing values

```python
without_filling = titianic_data.pivot_table(index="pclass", values="fare", aggfunc="mean", fill_value=0)
print(without_filling)
```

7. Margins (totals)

```python
with_margin = titianic_data.pivot_table(index="pclass", values="fare", aggfunc="mean", margins=True)
print(with_margin)
```

```python
port_status = titianic_data.pivot_table(index="embarked", values={"fare", "survived"}, aggfunc= numpy.sum)
print(port_status)
```

# DataFrame.loc[]

What it does:
- Sorts the DataFrame by age
- ascending=False → highest age first (descending order)
- inplace=False → creates a new DataFrame (does NOT modify original)

👉 So:
new_titianic_data = sorted version of your data (oldest → youngest)

```python
new_titianic_data = titianic_data.sort_values("age", inplace=False, ascending=False)
new_titianic_data
```

🧠 What is .loc?

👉 .loc is used to select rows (and columns) by LABEL (index name)

| Expression | Meaning                    |
| ---------- | -------------------------- |
| `.loc[0]`  | row with index = 0         |
| `.iloc[0]` | first row (position-based) |


🔥 Example

After sorting:

index   name    age
2       Bob     60
0       John    45
1       Alice   30

👉 .loc[0]
John    45

👉 .iloc[0]
Bob     60


 🧠 Final takeaway
.loc[] → uses index label
.iloc[] → uses position
Sorting does NOT change index unless you reset it

🧠 What this code does (simple idea)

👉 It counts how many missing values (NaN) are present in each column of your DataFrame.

⚡ Easier way (recommended)

You don’t actually need a custom function. Pandas already has this:

titianic_data.isnull().sum()


👉 Same result, simpler and faster 👍

🧠 Final takeaway
apply() → runs function on each column
pd.isnull() → finds missing values
Your code → counts nulls column-wise

🔁 What .apply() does
column_null_count = titianic_data.apply(null_count)


👉 .apply():

Runs null_count() on each column
Collects results

```python
import pandas as pd

def null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)
```

```python
column_null_count = titianic_data.apply(null_count)
print(column_null_count)
```

# DataFrame.apply(function, axis=1)

```python
def is_minor(row):
    if row["age"] < 18:
        return True
    else:
        return False
```

```python
minors = titianic_data.apply(is_minor, axis=1)
minors
```

```python
def which_class(row):
    pclass = row['pclass']

    if pd.isnull(pclass):
        return "Unknown"
    elif pclass == 1:
        return '1st class'
    elif pclass == 2:
        return '2nd class'
    elif pclass == 3:
        return '3rd class'
    
```

```python
with_class = titianic_data.apply(which_class, axis=1)
with_class
```
