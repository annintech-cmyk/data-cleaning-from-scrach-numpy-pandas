```python
in_g = "(g)"
in_mg = "(mg)"
in_kcal = "Kcal"
in_mcg = "(mcg)"
```

```python
def get_spec(spec):
    col_names = food_info.columns.tolist()
    gram_columns =[]

    for c in col_names:
        if c.endswith(spec):
            gram_columns.append(c)



    gram_df = food_info[gram_columns]
    return gram_df.head(3)
    
```
