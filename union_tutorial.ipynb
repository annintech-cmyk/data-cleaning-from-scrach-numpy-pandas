Union

```python
import pandas as pd

df1 = pd.DataFrame({
    "id": [1, 2],
    "name": ["Alice", "Bob"]
})

df2 = pd.DataFrame({
    "id": [3, 2],
    "name": ["Carol", "Bob"]
})
```

🚀 UNION ALL (Keep duplicates)

```python
result = pd.concat([df1, df2])
print(result)
```

⭐ Real SQL UNION (Remove duplicates)

```python
result = pd.concat([df1, df2]).drop_duplicates()
print(result)
```

Reset index after UNION (clean dataset)

```python
result = pd.concat([df1, df2]).drop_duplicates().reset_index(drop=True)
print(result)
```

| SQL       | Pandas                          |
| --------- | ------------------------------- |
| UNION ALL | `pd.concat()`                   |
| UNION     | `pd.concat().drop_duplicates()` |
| JOIN      | `pd.merge()`                    |
