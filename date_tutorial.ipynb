```python
import pandas as pd

# Example string
date_string = '2026-03-21 14:30:00'

# Convert string to datetime
date_time_obj = pd.to_datetime(date_string)
print(date_time_obj)
```

Extracting Year, Month, and Day

```python
year= date_time_obj.year
month=date_time_obj.month
day=date_time_obj.day
# Extract year, month, and day
print("Year:", date_time_obj.year)
print("Month:", date_time_obj.month)
print("Day:", date_time_obj.day)
print(date_time_obj.day_of_week)
print(date_time_obj.weekofyear)
```

```python
# Creating a DataFrame with missing datetime data
data = {'date': ['2026-11-20 14:30:00',  None, None, '2026-11-23 10:00:00']}
df = pd.DataFrame(data)

# Convert to datetime
df['date'] = pd.to_datetime(df['date'])

print(df)
```

Example: Adding Days to a Date

```python
from pandas import Timedelta

# Add 5 days to a datetime object
new_date = date_time_obj + Timedelta(days=5)
print("New Date:", new_date)
```

Example: Converting Time Zone


```python
# Set time zone
if date_time_obj.tzinfo is None:
    date_time_obj = date_time_obj.tz_localize('UTC')

date_time_obj_est = date_time_obj.tz_convert('US/Eastern')
print("Converted Time Zone:", date_time_obj_est)

```
